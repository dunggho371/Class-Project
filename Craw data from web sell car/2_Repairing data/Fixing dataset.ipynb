```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv('fixed_data.csv')
df
```

```python
# Add index
df = df.reset_index()
```

```python
data= df.query('verified == "FALSE"|verified == "TRUE"')
result = df.query('index not in @data.index')
result
```

```python
result['price_comparison_position']=result['verified']
result
```

```python
import numpy as np
result['verified']= np.nan
result
```

```python
df = result.combine_first(df)
```

```python
df['verified'].unique
```

```python
data2= df.query('map_item_attributes == "FALSE"|map_item_attributes == "TRUE"')
data2
```

```python
data3=data2.copy()
data3['map_item_attributes']=data2['whatsapp_integration_available']
data3['whatsapp_integration_available']=data2['map_item_attributes']
```

```python
df = data3.combine_first(df)
```

```python
# Extracting the values
df['BRAND'] = df['map_item_attributes'].str.extract(r"'name': 'Marca', 'value_id': '.*?', 'value_name': '(.*?)'")
df['MODEL'] = df['map_item_attributes'].str.extract(r"'name': 'Modelo', 'value_id': '.*?', 'value_name': '(.*?)'")
df['VEHICLE_YEAR'] = df['map_item_attributes'].str.extract(r"'name': 'A.*?o', 'value_id': '.*?', 'value_name': '(.*?)'")

df
```
