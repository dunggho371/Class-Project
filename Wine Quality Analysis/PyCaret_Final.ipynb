```python
import pandas as pd
from pycaret.regression import *
import time
import psutil

# Ghi nhận thời gian bắt đầu
start_time = time.time()

# Tải dữ liệu ví dụ (ví dụ: wine)
train_data = pd.read_csv('/kaggle/input/2023sumail303m/train.csv', sep=';')
```

```python
# Thiết lập PyCaret và huấn luyện mô hình trên tập train
reg = setup(data=train_data, target='quality', normalize=True)
```

```python
best = compare_models()
```

```python
finalize_model(best)
```

```python
evaluate_model(best)
```

```python
# Đường dẫn đến tập dữ liệu mới
test_data = pd.read_csv("/kaggle/input/2023sumail303m/test.csv", sep=';')

# Bỏ cột 'id' trong tập test
test_data_processed = test_data.drop('id', axis=1)
```

```python
# Dự đoán 'quality' trên tập test đã được xử lý
predictions = predict_model(best, test_data_processed)

# Tạo DataFrame mới chứa 'id' và 'quality'
output_df = pd.DataFrame({'id': test_data['id'], 'quality': predictions['prediction_label']})
output_df
```

```python
output_df.to_csv('Pycaret_Final.csv',index= False)
```

```python
# Ghi nhận thời gian kết thúc
end_time = time.time()
execution_time = end_time - start_time
```

```python
# Tính toán thông tin hiệu suất
process = psutil.Process()
memory_usage = process.memory_info().rss / 1024 ** 2  # Đơn vị: Megabytes
```

```python
# In ra thông tin
print('-' * 20)
print('Model tốt nhất:', best)
print('Thời gian thực thi:', execution_time, 'giây')
print('Sử dụng bộ nhớ:', memory_usage, 'MB')
```
